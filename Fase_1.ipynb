Gestion de Nulos en Columnas `Cancellation Year` / `Cancellation Month`





* Observamos que los nulos indican Activo
y los que tienen año indican el año que se dieron de baja

* Creamos una columna Status , en posicion anterior a Cancellation Year y Cancellation Month que indique el estado del cliente 

* No eliminamos la columnas con nulos para futuros estudios

```python
# Creamos la columna 'Status': 
# Si 'Cancellation Year' es nulo (isna), ponemos 'Activo'
# Si tiene valor (no es nulo), ponemos 'Inactivo'

df_final_copia['Status'] = np.where(
    df_final_copia['Cancellation Year'].isna(), 
    'Activo', 
    'Inactivo'
)

# Verificamos cuántos hay de cada uno para confirmar el 87.66%
print(df_final_copia['Status'].value_counts(normalize=True) * 100)
```

```python
# para que se entienda organizo la columnas , poniendo primero 'Status'
#  Obtenemos la lista actual de columnas
cols = df_final_copia.columns.tolist()


idx = cols.index('Cancellation Year')

# Movemos 'Status' (que está al final) a esa posición
# Primero la quitamos de su sitio actual
cols.remove('Status')
# Luego la insertamos en la posición deseada
cols.insert(idx, 'Status')
# Reindexamos el DataFrame con el nuevo orden
df_final_copia = df_final_copia[cols]
```
